```python
import os
import gym

# ImportRL algorithm such as SAC/PPO/A2C/DDPG/DQN/HER/TD3
from stable_baselines3 import PPO

# vectorize the environment
from stable_baselines3.common.vec_env import DummyVecEnv

# To make easy to evaluate the model/ average reward..
from stable_baselines3.common.evaluation import evaluate_policy
```

# load environment

```python
environment_name = 'CartPole-v0'
env = gym.make(environment_name)
```

```python
# Think of an episode as one full game within the environment.
# Some environments have a fixed episode length e.g CartPole which is 200 frames.
# others are continueous, e.g Breakout, play until you run out of lives.
episodes = 5
for episode in range (1, episodes+1):
    state = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score += reward
    print('Episode:{} Score:{}'.format(episode, score))
# env.close()
```

# Train an RL model

```python
# Mak your directories first
log_path = os.path.join('Training','Logs')
log_path
```

```python
# pip install torch==1.9.0+cu102 torchvision==0.10.0+cu102 torchaudio===0.9.0 -f https://download.pytorch.org/whl/torch_stable.html
```

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env])

# Benefit of using stable based, here we are using Multilayer perceptron policy (neural network), we have other options such as LSTM,CNN
# Traiding , finance , gaming app are not available in stable policies
# verbos=1 because we want to look out the result
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
# tensorboard_log = log_path  ==>  specify a logging directory when you initialize your model, this is because we will be able to monitore tensorboard
# stable-baselines3 has 3 policy types: MlpPolicy, CnnPolicy, MultiInputPolicy
```

```python
PPO??
# Model is not an environment
```

```python
# The complixity of the environment has gonna define how many timestes you need to train the model
model.learn(total_timesteps=20000)
```

# Save and reload model

```python
PPO_path = os.path.join('Training', 'Saved Models', 'PPO_model_catrpole')
```

```python
#To remove the model 
del model
model.learn(total_timesteps=100)
```

```python
# To reload the model
model = PPO.load(PPO_path, env=env)

# Just for check
model.learn(total_timesteps=100)
```

# Evaluation

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)

# average reward is 200 ==> great, it solved
# standard deviasion =0
```

# Test model

```python
# To deploy the model, we have to observe the environment and take action based on observation, not only sample action. => change action to observation
episodes = 5
for episode in range (1, episodes+1):
    obs = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render()
        # instead of "action = env.action_space.sample()" we must predict based on observation
        action, _ = model.predict(obs)  #USING MODEL HERE
        obs, reward, done, info = env.step(action)
        score += reward
    print('Episode:{} Score:{}'.format(episode, score))
# env.close()
```

# Adding a callback to thr training stage

```python
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
```

```python
save_path = os.path.join('Training', 'Saved Model')
```

```python
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=200, verbose=1)
eval_callback = EvalCallback(env,
                            callback_on_new_best=stop_callback,
                            eval_freq=10000,
                            best_model_save_path=save_path,
                            verbose=1)
```

```python
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=20000, callback=eval_callback)
```

# Changing policies (changing NN)

```python
# 4 layers NN with 128 units
net_arch = [dict(pi=[128,128,128,128], vf=[128,128,128,128])]
```

```python
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path, policy_kwargs={'net_arch':net_arch})
```

```python
model.learn(total_timesteps=20000, callback=eval_callback)
```
